## Eliminar valores faltantes

```python
messiLimpioDf = messiLimpioDf.dropna(how='all') # Drop rows that has ALL Nan Values
messiLimpioDf.tail()
```

```python
messiLimpioDf.shape
```

```python
messiLimpioDf.isnull().sum()*100/len(messiLimpioDf) # porcentajes de valores nulos por columna post limpieza total de nulos
```

### Eliminar filas que tienen un porcentaje de Nan mayor al 80%

```python
porcentajeNulos = 80.0
minValorNoNulos = int((100-porcentajeNulos)*messiLimpioDf.shape[1]/100)
messiLimpioDf = messiLimpioDf.dropna(axis=0, thresh=minValorNoNulos)
messiLimpioDf.info()
# sería lo mismo hacer
# messiLimpioDf = messiLimpioDf.dropna(axis = 0, thresh = int(messiLimpioDf.shape[1]*0.2))
```

## Eliminar columnas compeltas que no sirven para el análisis o no tienen info relevante

```python
messiLimpioDf = messiLimpioDf.drop(axis=1, columns='Pareja')
messiLimpioDf.shape
```

```python
messiLimpioDf.head(3)
```

## Eliminar columnas duplicadas

```python
messiLimpioDf = messiLimpioDf.T.drop_duplicates().T  # this removes all columns that have the same data regardless of column names
messiLimpioDf.shape
```

## Separar datos útiles para el anbálisis. En este caso el año

```python
messiLimpioDf['Año'] = messiLimpioDf['Fecha'].dt.strftime('%Y')
messiLimpioDf.sample(5)
```

## Separar columna Resultado en: Resultado_equipo y Resultado_oponente 

```python
resultadoSeparado = messiLimpioDf['Resultado'].str.split(':',n = 1, expand=True) #the expand parameter is set to True, which means it will return a data frame with all separated strings in different columns
#print(resultadoSeparado[1])
messiLimpioDf['Resultado_equipo'] = np.where(messiLimpioDf['Evento']=='Home', resultadoSeparado[0], resultadoSeparado[1]) #condition -When True, yield x, otherwise yield y.
messiLimpioDf['Resultado_oponente'] = np.where(messiLimpioDf['Evento']=='Away', resultadoSeparado[0], resultadoSeparado[1])
messiLimpioDf.sample(10)
```

```python
messiLimpioDf.dtypes
```

## Convertir columna minuto en tipo numérico

```python
prueba = messiLimpioDf  # elimiando todos los " ' " de la columna "Minutos"
prueba["Minuto"] = prueba["Minuto"].replace("'","",regex=True)
prueba["Minuto"]
```

### Separar columna minuto en columnas minuto_partido y minuto_extra

```python
minutosSeparada = messiLimpioDf['Minuto'].str.split('+',n = 1, expand=True) #the expand parameter is set to True, which means it will return a data frame with all separated strings in different columns
minutosSeparada
```

```python
messiLimpioDf['minuto_partido'] = minutosSeparada[0]
messiLimpioDf['minuto_extra'] = minutosSeparada[1]
messiLimpioDf.sample(5)
```

## ¿Mete más goles en el primer o en el segundo tiempo?

```python
contadorPrimerT=0
contadorSegundoT=0

for minuto in messiLimpioDf['minuto_partido']:
  if minuto<=45:
    contadorPrimerT += 1
  else:
    contadorSegundoT += 1
golesTotales = contadorPrimerT + contadorSegundoT
print('Porcentaje de goles en el primer tiempo: ' + str(contadorPrimerT*100/golesTotales))
print('Porcentaje de goles en el segundo tiempo: ' + str(contadorSegundoT*100/golesTotales))
```

```python
if contadorPrimerT>contadorSegundoT:
  print('Messi mete más goles en el primer tiempo')
else:
  print('Messi mete más goles en el segundo tiempo')
```
